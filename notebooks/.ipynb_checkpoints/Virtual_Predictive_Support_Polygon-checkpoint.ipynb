# Virtual Predictive Support Polygon

**Important**: Use _nbviewer_ to look at this notebook. Github has some problems with rendering equations properly: [open this notebook in nbviewer](https://nbviewer.jupyter.org/github/ThomasSchnapka/quaro/blob/master/notebooks/Virtual_Predictive_Support_Polygon.ipynb)

This notebook exemplifies the use of the concept of a *Virtual Predictive Support Polygon* as described in the [MIT Cheetah 3 paper](https://dspace.mit.edu/bitstream/handle/1721.1/126619/IROS.pdf?sequence=2&isAllowed=y)

The concept is used to calculate a suitable COM location for quadruped robots depending on the leg schedule. Using the knowledge of the leg schedule, the COM gets moved away from legs in swing phase and towards legs in support phase.

Remark: In order to simplify the use of indices for calculating with values of adjacent leg and adapting it to the conventions used in the paper, this notebook uses a _different leg numbering_ than usual. The leg are numerated counter-clockwise looking at the +z-direction: LF = 0, HL = 1, HR = 2, FL = 3

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from scipy import special

# Basic definitions
support_ratio = 0.8
cycle_time = 1.5
phase = np.array([0, 0.25, 0.5, 0.75])

def get_leg_states(t, cycle_time, support_ratio, phase):
    '''returns a bool array with leg states (1 = supporting, 0 = swinging)'''
    state = (phase + t)/cycle_time
    phi = state%1
    state = phi <= support_ratio
    # in this paper, phi goes from 0 to 1 in both phases
    phi[state] /= support_ratio
    phi[~state] = (phi[~state]-support_ratio)/(1-support_ratio)
    return phi, state
```

### A Probablistic view at contact

The paper introduces weighting factors representing the probability for the leg having contact with the ground during either stance ($K_{c \phi}$) or swing ($K_{\overline{c} \phi}$) phase:

$$
K_{c \phi} = \frac{1}{2}\left(\text{erf} \left(\frac{\phi}{\sigma_{c0}\sqrt{2}}\right) + \text{erf} \left(\frac{1 - \phi}{\sigma_{c1}\sqrt{2}}\right)
\right)
$$
 
$$
K_{\overline{c} \phi} = \frac{1}{2}\left(
2 + \text{erf} \left(\frac{- \phi}{\sigma_{\overline{c}0}\sqrt{2}}\right) + \text{erf} \left(\frac{\phi - 1}{\sigma_{\overline{c}1}\sqrt{2}}\right)
\right)
$$

Remark: there two different variables $phi$ in the paper: $\varphi$ denotes the normalized time per leg while $\Phi$ is the total weighting factor.
In the following code, $\Phi$ is described as `phi_total`

```python
# calculate weighting factors
def get_weighting_factor(phi, sigma0, sigma1, sigma0_, sigma1_):
    '''returns both weighting factors given the leg phase'''
    Kc  = 0.5*(   special.erf(phi  / (sigma0*np.sqrt(2)))
                + special.erf((1-phi) / (sigma1*np.sqrt(2))))
    Kc_ = 0.5*(2 + special.erf(-phi / (sigma0_*np.sqrt(2)))
                 + special.erf((phi-1) / (sigma1_*np.sqrt(2))))
    return Kc, Kc_


def get_phi_total(phi, leg_state, sigma0=0.1, sigma1=0.1,
                                  sigma0_=0.1,sigma1_=0.1):
    '''return total weighting factor for a single leg'''
    Kc, Kc_ = get_weighting_factor(phi, sigma0, sigma1, sigma0_, sigma1_)
    phi_total = leg_state*Kc + ~leg_state*Kc_
    return phi_total
```

```python
N = 100
t = np.linspace(0, cycle_time, N)

K = np.zeros((N, 2))
phi_total = np.zeros(N)
for i in range(N):
    phi, state = get_leg_states(t[i], cycle_time, support_ratio, phase)
    K[i, 0], K[i, 1] = get_weighting_factor(phi[0], 0.1, 0.1, 0.1, 0.1)
    phi_total[i] = get_phi_total(phi[0], state[0], 0.1, 0.1, 0.1, 0.1)

plt.plot(t, K[:,0], "--", label="$K_c$")
plt.plot(t, K[:,1], "--", label="$K_{\overline{c}}$")
plt.plot(t, phi_total, label="$\Phi$")
plt.xlabel("t [s] for a whole cycle")

plt.legend()
plt.show()
```

The above plot can give an intuitive understanding for the introduced factors.
 $K_c$ and $K_{\overline{c}}$ represent the probability that the corresponding leg is capable of supporting the weight in the respective phase. $\Phi$ switches between those factors based on the current state and thus gives an overall probability for contact.

### Virtual Points

Using the phase-based weighting factors a set of virtual points is defined. These weight the ability of each individual leg to support the COM compared to the adjacent legs. The virtual points slide on a line between a foot and its neighbors and are given by:

$$
\left( \begin{array}{c} \xi_i^- \\ \xi_i^+ \end{array} \right)
= \left(\begin{array}{rr}  
\textbf{p}_i & \textbf{p}_{i-} \\ 
\textbf{p}_i & \textbf{p}_{i+} \\ 
\end{array}\right)
\left( \begin{array}{c} \Phi_i \\ 1-\Phi_i \end{array} \right)
$$

The + and - sign reference the adjacent leg (clockwise -, counterclockwise +)

```python
def get_virtual_points(phi_total, p, p_cw, p_ccw):
    '''returns two virtual points per leg'''
    xi_cw  = p*phi_total +  p_cw*(1-phi_total)
    xi_ccw = p*phi_total + p_ccw*(1-phi_total)
    return xi_cw, xi_ccw
```

The virtual points are used to calculate the predictive support polygon vertex:

$$
\xi_i = \frac{\phi_i\textbf{p}_i + \phi_{i-}\xi_i^- + \phi_{i+}\xi_i^+}{\phi_{i} + \phi_{i-} + \phi_{i+}}
$$

Now we can calculate the desired COM position using the averade of all predictive support polygon vertices:

$$
\textbf{p}_{COM,d} = \frac{1}{N} \sum\xi_i
$$
    

```python
def get_vertex(phi_total, phi_total_cw, phi_total_ccw, p, xi_cw, xi_ccw):
    '''return predictive support polygon vertex for a single foot'''
    return ((phi_total*p + phi_total_cw*xi_cw + phi_total_ccw*xi_ccw)
            /(phi_total + phi_total_cw + phi_total_ccw))

def get_com(vertices):
    '''return COM position by averaging vertex matrix'''
    return np.mean(vertices, axis=0)
```

Lets make a test using 2D leg coordinates:

```python
# leg coordinates
p = np.array([[0.09, 0.04], [-0.09, 0.04], [-0.09, -0.04], [0.09, -0.04]])
t = 2.2

phi, leg_states = get_leg_states(t, cycle_time, support_ratio, phase)

# calculate phi_total and individual xi values
phi_total = np.zeros(4)
xi_cw  = np.zeros((4,2))
xi_ccw = np.zeros((4,2))
for i in range(4):
    phi_total[i] = get_phi_total(phi[i], leg_states[i])
    xi_cw[i], xi_ccw[i] = get_virtual_points(phi_total[i],
                                             p[i], p[i-1], p[(i+1)%4])
    
# calucate the predictive support polygon vertices
vertices = np.zeros((4,2))
for i in range(4):
    vertices[i] = get_vertex(phi_total[i], phi_total[i-1], phi_total[(i+1)%4],
                             p[i], xi_cw[i], xi_ccw[i])
# resulting desired COM
com = get_com(vertices)
```

Plot the results:

```python

# plot result colored indiviually for each leg
colors = ["r", "b", "g", "k"]
for i in range(4):
    plt.plot(p[i,0], p[i,1], "o", color=colors[i])
    plt.plot(vertices[i, 0], vertices[i, 1], "s", color=colors[i])
    plt.plot(xi_cw[i, 0], xi_cw[i, 1], "_", color=colors[i])
    plt.plot(xi_ccw[i, 0], xi_ccw[i, 1], "+", color=colors[i])
    
# plot polygons
plt.plot(np.append(p[:,0], p[0,0]), np.append(p[:,1], p[0,1]),
         "--", color="gray", alpha=0.1)
plt.plot(np.append(vertices[:,0], vertices[0,0]), 
         np.append(vertices[:,1], vertices[0,1]),
         "--", color="blue", alpha=0.1)

# show which legs are supporting 
plt.plot(p[leg_states, 0], p[leg_states, 1], ".", alpha=0.1, ms=70)

# plot COM
plt.plot(com[0], com[1], ".")
plt.xlim([0.13, -0.13])
plt.ylim([0.07, -0.07])
plt.show()

print("COM =", com)
```

What do we see in this plot?

All markers in one color belong to a single leg. The circles denote
the foot position, the - and + signs are the calculated $\xi$ values and
the squares are the resulting weighted vertices. A gray shade marks the legs, that are in supporting state. The result of this whole
calculation, the COM, can be seen as a dot in the middle of the plot.

### Parameters

As we can see in the plot above, the COM moves just slightly. The reasons for this might are the values we have chosen for the different 𝜎σ. In the following we want to exploit the influences of those parameters.

Firstly, we write a function that plots the COM position depending on different sigma values:

```python
def plot_com(ax0, axx, axz, sigma0, sigma1, sigma0_, sigma1_):
    N = 100
    com = np.zeros((N,2))
    tsteps = np.linspace(0, cycle_time, N)
    # uncomment this for other phases
    # phase = np.array([0.5, 0, 0.5, 0]) # trot
    # phase = np.array([0.05, 0.25, 0.75, 0.55]) # crawl
    
    for n in range(N):
        phi, leg_states = get_leg_states(tsteps[n], cycle_time,
                                         support_ratio, phase)
        
        # calculate phi_total and individual xi values
        phi_total = np.zeros(4)
        xi_cw  = np.zeros((4,2))
        xi_ccw = np.zeros((4,2))
        for i in range(4):
            phi_total[i] = get_phi_total(phi[i], leg_states[i],
                                         sigma0, sigma1, sigma0_, sigma1_)
            xi_cw[i], xi_ccw[i] = get_virtual_points(phi_total[i],
                                                     p[i], p[i-1], p[(i+1)%4])
        # calucate the predictive support polygon vertices
        vertices = np.zeros((4,2))
        for i in range(4):
            vertices[i] = get_vertex(phi_total[i], phi_total[i-1], 
                                     phi_total[(i+1)%4], p[i], 
                                     xi_cw[i], xi_ccw[i])
        com[n] = get_com(vertices)
        
    label =  "$\sigma$ = " + str(sigma0) + ", " + str(sigma1) + ", " \
            + str(sigma0_) + ", " + str(sigma1_)
    ax0.plot(com[:,0], com[:,1], label=label)
    axx.plot(tsteps, com[:,0])
    axy.plot(tsteps, com[:,1])
```

Now we can plot the COM trajectory for differend $\sigma$ values

```python
fig = plt.figure(figsize=(9,7))
gs = fig.add_gridspec(2, 3)
ax0 = plt.subplot(gs[:,1:])
axx = plt.subplot(gs[0,0])
axy = plt.subplot(gs[1,0], sharex=axx)
plot_com(ax0, axx, axy, 0.1, 0.1, 0.1, 0.1)
plot_com(ax0, axx, axy, 0.5, 0.5, 0.5, 0.5)
plot_com(ax0, axx, axy, 0.5, 0.5, 0.1, 0.1)
plot_com(ax0, axx, axy, 0.1, 0.1, 0.5, 0.5)
plot_com(ax0, axx, axy, 0.5, 0.5, 0.5, 0.1)
ax0.plot(p[:,0], p[:,1], "o", c="k")
ax0.text(p[0,0]-0.01, p[0,1], "0 /LF")
ax0.text(p[1,0]-0.01, p[1,1], "1 /HL")
ax0.text(p[2,0]-0.01, p[2,1], "2 /HR")
ax0.text(p[3,0]-0.01, p[3,1], "3 /FR")
ax0.legend(bbox_to_anchor=(0.5,-0.1))
ax0.set_title("COM in x/y")
ax0.set_xlabel("x")
ax0.set_xlim([0.13, -0.13])
ax0.set_ylim([0.07, -0.07])
axx.set_title("COM in x/t and y/t")
axx.set_ylabel("x")
axy.set_ylabel("y")
axy.set_xlabel("t")
plt.suptitle("COM movement for a phase of " + str(phase))
plt.show()
```

What do we learn from this?
- the lower all $\sigma$ values are, the bigger the COM movement
- small values for $\sigma_c$ and high values for $\sigma_{\overline{c}}$ result in a smoother COM trajectory than in the inverted case, which is important to minimize unnecessary accelerations
- changing single $\sigma$ values instead of whole pairs (no matter which pair) result in a earlier movement in the corresponding direction (e.g. in the plot in the cw direction)

### Conclusion

Trying out several phase configurations, the method of _Virtual Predictive Support Polygons_ yield promising COM trajectories. The next step is to implement this method into the current robot software and make some experiments with real hardware. In order to do this, a more efficient calculation must be developed as the used nested Python loops are too slow for real time applications.
